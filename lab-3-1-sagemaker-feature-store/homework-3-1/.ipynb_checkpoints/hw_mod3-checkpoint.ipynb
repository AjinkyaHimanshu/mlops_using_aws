```python
import pandas as pd
import numpy as np
from datetime import datetime, timezone

# ---------------------------------------------------------
# Step 1: Merge housing data with maps data
# ---------------------------------------------------------
df = pd.merge(
    house_df,
    maps_df[['longitude', 'latitude', 'neighborhood-political', 'postal_code']],
    on=['longitude', 'latitude'],
    how='left'
)

# ---------------------------------------------------------
# Step 2: Primary Key (neighborhood)
# ---------------------------------------------------------
df['neighborhood'] = df['neighborhood-political']

# ---------------------------------------------------------
# Step 3: Event Time
# ---------------------------------------------------------
df['event_time'] = datetime.now(timezone.utc).strftime("%Y-%m-%dT%H:%M:%S.%fZ")

# ---------------------------------------------------------
# Step 4: Ocean Proximity One-Hot Encoding
# ---------------------------------------------------------
# ['NEAR BAY', '<1H OCEAN', 'INLAND', 'NEAR OCEAN', 'ISLAND']
df['<1h ocean'] = (df['ocean_proximity'] == '<1H OCEAN').astype(int)
df['inland'] = (df['ocean_proximity'] == 'INLAND').astype(int)
df['island'] = (df['ocean_proximity'] == 'ISLAND').astype(int)
df['near bay'] = (df['ocean_proximity'] == 'NEAR BAY').astype(int)
df['near ocean'] = (df['ocean_proximity'] == 'NEAR OCEAN').astype(int)

# ---------------------------------------------------------
# Step 5: Median House Value (avg per neighborhood, capped at 500k)
# ---------------------------------------------------------
df['median_house_value'] = df.groupby('neighborhood')['median_house_value'].transform('mean')
df['median_house_value'] = df['median_house_value'].clip(upper=500000).round(5)

# ---------------------------------------------------------
# Step 6: Median House Age (avg per neighborhood, discretized into 10-year bins)
# ---------------------------------------------------------
df['median_house_age'] = df.groupby('neighborhood')['housing_median_age'].transform('mean')
df['median_house_age_bin'] = (df['median_house_age'] // 10) * 10
df['median_house_age_bin'] = df['median_house_age_bin'].fillna(0).astype('int64')

# ---------------------------------------------------------
# Step 7: Total Households (avg per neighborhood, rounded up)
# ---------------------------------------------------------
df['total_households'] = df.groupby('neighborhood')['households'].transform('mean')

# Handle missing values before converting to int
df['total_households'] = df['total_households'].fillna(0)   # replace NaN with 0
df['total_households'] = np.ceil(df['total_households']).astype(int)

# ---------------------------------------------------------
# Step 8: Bedrooms per Household (derived, imputed by postal_code average)
# ---------------------------------------------------------
df['bedrooms_per_household'] = df['total_bedrooms'] / df['households']
df['bedrooms_per_household'] = df['bedrooms_per_household'].round(5)

# Impute missing values using postal_code averages
df['bedrooms_per_household'] = df.groupby('postal_code')['bedrooms_per_household'].transform(
    lambda x: x.fillna(x.mean()).round(5)
)

# ---------------------------------------------------------
# Step 9: Final Feature Group Schema
# ---------------------------------------------------------
neighborhood_features = df[[
    'neighborhood',
    'event_time',
    '<1h ocean',
    'inland',
    'island',
    'near bay',
    'near ocean',
    'median_house_value',
    'median_house_age_bin',
    'total_households',
    'bedrooms_per_household'
]].drop_duplicates(subset=['neighborhood'])

neighborhood_features = neighborhood_features.dropna(axis=0)

neighborhood_features = neighborhood_features.rename(
    columns={"<1h ocean": "less_than_one_hour_ocean", 'near bay': 'near_bay', 'near ocean': 'near_ocean'}) 
```

```python
def cast_object_to_string(data_frame):
    for label in data_frame.columns:
        if data_frame.dtypes[label] == "object":
            data_frame[label] = data_frame[label].astype("str").astype("string")

# cast object dtype to string. The SageMaker FeatureStore Python SDK will then map the string dtype to String feature type.
cast_object_to_string(neighborhood_features)
```

```python
from sagemaker.feature_store.feature_definition import FeatureDefinition, FeatureTypeEnum
from sagemaker.feature_store.feature_group import FeatureGroup
import time
# ---------------------------------------------------------
# Step 1: Define the schema for the Neighborhood Feature Group
# ---------------------------------------------------------
feature_definitions = [
    FeatureDefinition(feature_name="neighborhood", feature_type=FeatureTypeEnum.STRING),
    FeatureDefinition(feature_name="event_time", feature_type=FeatureTypeEnum.STRING),
    FeatureDefinition(feature_name="less_than_one_hour_ocean", feature_type=FeatureTypeEnum.INTEGRAL),
    FeatureDefinition(feature_name="inland", feature_type=FeatureTypeEnum.INTEGRAL),
    FeatureDefinition(feature_name="island", feature_type=FeatureTypeEnum.INTEGRAL),
    FeatureDefinition(feature_name="near_bay", feature_type=FeatureTypeEnum.INTEGRAL),
    FeatureDefinition(feature_name="near_ocean", feature_type=FeatureTypeEnum.INTEGRAL),
    FeatureDefinition(feature_name="median_house_value", feature_type=FeatureTypeEnum.FRACTIONAL),
    FeatureDefinition(feature_name="median_house_age_bin", feature_type=FeatureTypeEnum.INTEGRAL),
    FeatureDefinition(feature_name="total_households", feature_type=FeatureTypeEnum.INTEGRAL),
    FeatureDefinition(feature_name="bedrooms_per_household", feature_type=FeatureTypeEnum.FRACTIONAL),
]

# ---------------------------------------------------------
# Step 2: Create the FeatureGroup object
# ---------------------------------------------------------
feature_group_name = "NeighborhoodFeatureGroup"

neighborhood_feature_group = FeatureGroup(
    name=feature_group_name,
    sagemaker_session=feature_store_session,
    feature_definitions=feature_definitions
)

# ---------------------------------------------------------
# Step 3: Register the Feature Group in SageMaker
# Note: this matches the signature you showed (s3_uri + enable_online_store)
# ---------------------------------------------------------

def wait_for_feature_group_creation_complete(feature_group):
    status = feature_group.describe().get("FeatureGroupStatus")
    while status == "Creating":
        print("Waiting for Feature Group Creation")
        time.sleep(5)
        status = feature_group.describe().get("FeatureGroupStatus")
    if status != "Created":
        raise RuntimeError(f"Failed to create feature group {feature_group.name}")
    
    print(f"FeatureGroup {feature_group.name} successfully created.")
    time.sleep(20)
    
neighborhood_feature_group.create(
    s3_uri=f"s3://{default_s3_bucket_name}/{prefix}/{feature_group_name}",  # Offline store location
    record_identifier_name="neighborhood",  # Primary key
    event_time_feature_name="event_time",   # Event time column
    role_arn=role,                          # IAM role
    enable_online_store=True                # Enable Online Store
)

wait_for_feature_group_creation_complete(feature_group=neighborhood_feature_group)
```

```python
# ---------------------------------------------------------
# Step 4: Ingest engineered features into the Feature Store
# ---------------------------------------------------------
# records = neighborhood_features.to_dict(orient="records")

neighborhood_feature_group.ingest(
    data_frame=neighborhood_features,
    max_workers=3,   # parallel workers
    wait=True        # block until ingestion finishes
)
```

```python
def get_neighborhood_features(neighborhood_name):
    response = featurestore_runtime.get_record(
        FeatureGroupName="NeighborhoodFeatureGroup",
        RecordIdentifierValueAsString=neighborhood_name
    )
    return response
```
